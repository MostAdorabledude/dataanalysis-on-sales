```python
sales_data = pd.read_csv(r'C:\Users\Paul Fadayini\Desktop\titucodes\dataresources\Superstore_data.csv', encoding = 'latin1')
```

```python
sales_data['calculated_price']=sales_data['Sales']/ sales_data['Quantity']
```

```python
sales_data['calculated_price'].head()
```

```python
sales_data['Discounted_price']= sales_data['calculated_price']-(sales_data['Discount'] * sales_data['calculated_price'])
sales_data['Discounted_price'].head()
```

```python
sales_data.head()
```

```python
sales_data['Cost_price']=sales_data['calculated_price']-sales_data['Profit']
sales_data['Cost_price'].head()
```

```python
sales_data.head()
```

# TO FIND SALES BY CATEGORY OF PRODUCTS

```python
sales_by_cat=sales_data.groupby(sales_data['Category']).Sales.sum()
sales_by_cat.head()
```

# TO FIND THE PROFIT OF EACH CATEGORY OF PRODUCTS MADE

```python
profit_by_cat=sales_data.groupby(['Category']).Profit.sum()
print(profit_by_cat)
```

# sales by year

```python
#this code is to convert the data type in the 'ship date' from string to date time 
sales_data['Ship Date']=pd.to_datetime(sales_data['Ship Date'])
#after adusting data type, next line of code creates a column consisting of ust the year of shipment
sales_data['Year']=sales_data['Ship Date'].dt.year
sales_data.head()
```

```python
sales_by_year=sales_data.groupby(['Year'])['Sales']. sum()
print(sales_by_year)
```

# Top ten selling product

```python
#first group data by product name and total sales
top_ten=sales_data.groupby(['Product Name']).Sales.sum()
#then sort in ascending order to get the highest products sold
top_ten.sort_values(ascending=False).head(10)
```
